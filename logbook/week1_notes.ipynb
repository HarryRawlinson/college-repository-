versions and repositories:

Versions and Repos

When we work on code, we often make changes and improvements over time. Each significant change can be saved as a new version.

somtimes we want to be able to go back to previous versions if something goes wrong or if we want to see how the code has evolved.

the repostitory is where all these versions are stored.
although git doesnt store full versions of files it onlky stores the changes made (called delta) which saves space.
to save our work we have to commit and push changes to the repository. 

there are many was to commit in push within vs code.
one way is with the terminal using git commands.
the commands to commit and push are:
git add .
git commit -m "your message here"
git push origin main

you can also use the source control tab in vs code to stage, commit and push changes.
When we commit changes to a repository, its important to write meaningful commit messages that describe the changes made.
When we create a new project, we can initialize a Git repository in the project folder.


In the same folder (usually hidden), Git manages a repository that stores previous versions.

When a new version is ready (this might be changes in many files), we commit it to the repository.

(A repository is usually shortened to repo.)

USEFUL GIT COMMANDS:
git status - shows the current status of the repository, including changes that have been staged, unstaged, or untracked.
IDE's also provide visual indicators of these states.
it can also show what stages are committted and what changes are pending to github,

add . - stages all changes in the current directory for the next commit.

commit -m "message" - commits the staged changes to the repository with a descriptive message.

ipython is an enhanced version of the standard python interpreter that provides additional features such as syntax highlighting, tab completion, and magic commands. it is often used in data science and scientific computing communities.
it is similar to jupyter notebooks in that it allows for interactive coding and exploration of data.
it is more powerful than the standard interpreter and is worth exploring if you plan to do a lot of work in python although it may be overkill for simple tasks and not fit for someone just starting out with python.

coding concepts:
loop = a programming construct that allows for the repeated execution of a block of code until a certain condition is met.
while loop = a type of loop that continues to execute a block of code as long as a specified condition is true.
for loop = a type of loop that iterates over a sequence (such as a list or string) and executes a block of code for each item in the sequence.

print() = a built-in function in python that outputs text or other data to the console or terminal.
f string = a type of string in python that allows for the inclusion of variables and expressions within the string using curly braces {}.
print(f"Hello, {name}!") = an example of using an f-string to print a message that includes the value of the variable name.

variable = a data type which changes depending on input
constant = a data type which never changes

funcitons = a block of code that performs a specific task and can be reused multiple times throughout a program.
functions can take input parameters and return output values.
parameters = the input values that are passed to a function when it is called.
arguments = the actual values that are passed to a function when it is called.
