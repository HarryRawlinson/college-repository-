week 9 - 20th november 2025 - be dry not wet

dry = dont repeat yourself
wet = write everything twice (or waste everyone's time)

code gets read 10× more than written
humans can only hold ~20 lines in head
fix once → fixes everywhere
easier to test
easier to review
works better with copilot
teams can split work

rules
one function one job
copy-pasted code twice = make function already
functions max 20 lines (less is better)
4 spaces indentation only (not 1 not 2 not 8 not random)

typical program shape
stuff = read()
new_stuff = process(stuff)
write(new_stuff)

process() usually just loops and calls tiny function on each item

code smells
duplicated code
long parameter lists
anything you have to change in multiple places

lecturer showed some assignment code and cried (random indentation, giant 200-line main, same 15 lines copied 6 times)
