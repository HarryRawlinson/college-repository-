DATA TYPES
there are several data types in python including:

- integers (int): whole numbers, positive or negative, without decimals (e.g., 1, -5, 42)
- floating-point numbers (float): numbers with decimal points (e.g., 3.14, -0.001, 2.0)
often pythong will convert between these types automatically when performing operations.

- strings (str): sequences of characters enclosed in quotes (e.g., "hello", 'world', "123") 
number chars within strings cannot be used for mathematical operations unless converted to int or float.

- booleans (bool): represent truth values, either True or False.
boolean logic is often used in conditional statements such as if statements and loops. 

ARRAYS/LISTS
arrays/lists: ordered collections of items that can be of different data types. lists are mutable, meaning their contents can be changed after creation. lists are defined using square brackets [] and items are separated by commas.
example: my_list = [1, "hello", 3.14, True]

TUPLES
tuples: similar to lists but are immutable, meaning their contents cannot be changed after creation. tuples are defined using parentheses () and items are separated by commas.
example: my_tuple = (1, "hello", 3.14, True)
tuples can be converted to lists using the list() function and vice versa using the tuple() function.

sets 
sets: unordered collections of unique items. sets are mutable and defined using curly braces {} or the set() function.
example: my_set = {1, 2, 3} or my_set = set([1, 2, 3])
they have such usecases as removing duplicates from a list or performing mathematical set operations like union and intersection.

OFF THE HAPPY PATH
off the happy path: refers to situations where the code does not execute as expected due to errors or unexpected inputs. it is important to handle these situations gracefully to prevent crashes and ensure a good user experience.
there are 2 main philosophies for handling off the happy path scenarios:
look before you leap (LBYL): involves checking for potential errors or invalid inputs before executing code. this approach is proactive and aims to prevent errors from occurring in the first place.

ask for forgiveness not permission (EAFP): involves executing code and handling any errors that arise using try-except blocks. this approach is reactive and assumes that errors will occur, so it focuses on handling them when they do. 
there are pros and cons to each approach, and the choice between them often depends on the specific use

loops are used to repeat a block of code multiple times this can be handy in conjunction with off the happy path handling to ensure that code continues to run smoothly even when unexpected situations arise.

types of errors:
syntax errors: occur when the code violates the rules of the programming language. these errors are usually caught by the interpreter before the code is executed.
runtime errors: occur during the execution of the code and can be caused by various factors such as invalid inputs, division by zero, or file not found errors.
logical errors: occur when the code executes without crashing but produces incorrect results. these errors are often the most difficult to detect and fix as they do not generate error messages.

wet and dry code:
dry code: stands for "don't repeat yourself" and refers to the practice of avoiding code duplication. this can be achieved by using functions, classes, and modules to encapsulate reusable code. dry code is easier to maintain, as changes only need to be made in one place rather than multiple locations.
wet code: stands for "write everything twice" and refers to the practice of duplicating code. this can lead to maintenance issues, as changes need to be made in multiple locations, increasing the risk of errors and inconsistencies.



COMPUTATIONAL THINKING

computational thinking: a problem-solving approach that involves breaking down complex problems into smaller, more manageable parts. it includes concepts such as decomposition, pattern recognition, abstraction, and algorithm design.    

decomposition: breaking down a complex problem into smaller, more manageable parts.
think about what variables are needed and what their data types should be.

pattern recognition: identifying similarities or patterns in data or problems to help solve them more efficiently.
this is the stage where you think about where the code needs to iterate many times and what sort of loops are best 

abstraction: simplifying complex problems by focusing on the most important aspects and ignoring irrelevant details.
having a clear idea of the input and output of functions without worrying about the internal workings at first.


algorithm design: creating a step-by-step plan or set of instructions to solve a problem.
this can be done easily by writing pseudocode before actual coding to get a clear idea of the structure and flow of the program.

conclusions of copmutational thinking:
- helps in breaking down complex problems into manageable parts
- encourages efficient problem-solving through pattern recognition
- promotes clarity and simplicity through abstraction
- aids in creating structured solutions through algorithm design
can be used in everyday situations such as planning tasks, organizing information, and making decisions.
