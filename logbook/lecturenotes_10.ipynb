week 10 - 27th november 2025 - the joy of the command line

why command line still matters
- gui comes and goes, cli from 1970s still works exactly the same
- cloud servers, git, builds → all cli
- point-and-click git in vscode is slow and hides what actually happens

command line = text interface to computer
shell = program that reads your commands (usually bash or fish)

gui problems
- leaky abstractions (hides stuff you actually need to know)
- "my documents" who knows where it really is
- double-click opens with wrong app half the time

cli wins every time for programmer stuff
example tasks that are pain in gui but one liner in terminal

diff main_v1.py main_v2.py                  → shows exact differences
mkdir -p ~/focp-projects/project/src        → makes whole folder chain
rm log*.txt                                 → deletes 10 000 log files instantly
grep -rl "TODO" ~/focp-projects/project/src → lists every file with TODO
grep -rl "TODO" . | wc -l                   → counts how many files have TODO

proper git (not the vscode buttons)
cd focp-projects
git status
git add main.py
git commit -m "fix thing"
git push

basic commands he hammered
ls          → list files
cd          → change folder
pwd         → where am i
cp          → copy
mv          → move/rename
rm          → delete (no recycle bin)
mkdir       → make folder
cat         → dump file to screen
less        → view file page by page
head/tail   → first/last lines
grep        → search inside files
find        → search by name
which       → where is this program

conclusion slide literally said
"point-and-click is fine for grandmas, cli is for people who actually build stuff"
