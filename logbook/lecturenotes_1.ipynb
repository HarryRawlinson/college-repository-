what is a program ?
a program is a set of instructions that a computer can understand and execute so solve a speciific problem or perform a specific task

they are written in specified languages which is converted to machine code that the computer can understand
and then once compiled the computer can execute the instructions step by step
and produce the desired output
programming languages can be understood by humans and are designed to be easy to read and write

the 3 main parts of a program are 
sequence - the order in which instructions are executed 
selection - making decisions based on conditions
iteration - repeating a set of instructions until a certain condition is met
