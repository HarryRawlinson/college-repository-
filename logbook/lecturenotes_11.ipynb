jobs in industry (even in tiny 20-person companies)

DEV
write code, maintain code, fix bugs
most important skills: problem solving + communication + teamwork

TESTER
some peoples entire job is just breaking software and finding bugs

ANALYST / DESIGNER
talk to users, write requirements, plan future features

ARCHITECT
decide whole system layout, pick languages, platforms, databases etc

ADMINISTRATOR / SYSADMIN
run the servers, know linux inside out, keep everything alive

MANAGER
run agile projects, daily standups, keep everyone on track

DEVOPS
blurs dev + admin, everyone does a bit of everything
massive collaboration, you learn loads

SECURITY SPECIALIST
stop hackers, implement proper auth, pentest everything
