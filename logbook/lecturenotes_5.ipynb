week 5 - 23rd october 2025 - bugs (and ducks)
lecture about bugs and debugging and why everything is always broken

what even is a bug
a mistake in the code that makes the program do something weird or just crash
basically impossible to prove a program has zero bugs
you cant test every single possible situation ever
like today im using my laptop and phone in a way no one in history has ever done exactly the same
so yeah programs will always have bugs
best we can do is be pretty confident it works most of the time

two main types of errors
syntax errors - code is straight up illegal python, wont even run (spelling mistakes basically)
logic / semantic errors - code runs fine but does the wrong thing
  two kinds of logic errors:
  1. does what i wrote but i wrote the wrong thing (my fault)
  2. does what i wrote but the task was specified wrong (not really my fault lol)

bugs show up in two ways
program looks fine but gives wrong answer sometimes
program just crashes in some situations

some errors are super rare so we dont bother handling them properly
thats why windows gives you blue screen of death instead of something nice
not worth the effort to handle every crazy edge case
how to stop bugs happening so much
linus's law - "given enough eyeballs all bugs are shallow"
code reviews
no
 code goes into main without another human looking at it
even the senior devs have to do it
done with pull requests in git (branches → review → merge)

pair programming
two people one keyboard
one types one thinks out loud
basically nonstop code review
part of extreme programming (xp)
expensive but apparently really good for hard stuff

technical debt
when theres a quick dirty fix vs proper fix
everyone picks the quick dirty one then cries later
thats technical debt - you owe the code proper fix eventually

how to actually find bugs (debugging)
dry running - do it by hand on paper
use the actual debugger in the ide - step line by line watch variables
pepper with prints - just spam print() everywhere (fast and works surprisingly often
problem is remembering to delete them after

better prints - make a DEBUG = True at the top
then if DEBUG: print(stuff)
easy to turn off later

logging - proper programs write everything to a log file
can set levels (info warning error etc)

walk away / coffee break
your brain solves it when youre not looking at it
loads of people have the 10 minute rule - if stuck longer go drink something

explain to a friend
half the time you spot the bug while explaining it

rubber duck debugging (best one)
explain the code line by line to a rubber duck on your desk
you feel stupid but it actually works so well
you realise the mistake while talking to the duck
lecturer swears by this
