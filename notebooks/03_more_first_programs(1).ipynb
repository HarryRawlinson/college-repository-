_Suppose you purchase a 10 inch pizza. How many square *centimetres* of pizza do you have? Write a program below to find out. Remember that 10 inches is the diameter of the pizza._

```python
#import math
# input pizza inches 
#float(input("enter pizza diameter in inches"))
# go from inches to cm
#diameterinches = diametercm * 254 

#calculate area 
#pizzaareacm = diametercm * math.pi * (diameterinches ** 2)
# print output 
#print("the area of your pizza is" , round(pizzaareacm, 2), "square centimeters.")

import math

# Ask the user for the diameter in inches
diameterinches = float(input("Enter pizza diameter in inches: "))

# Convert diameter to centimeters
diametercm = diameterinches * 2.54

# Calculate radius in cm
radiuscm = diametercm / 2

# Calculate area in square centimeters
pizzaareacm = math.pi * (radiuscm ** 2)

# Print the result
print("The area of your pizza is", round(pizzaareacm, 2), "square centimeters.")
```

_Write a program where your user enters two integers, which represent weight measurements taken on two different days. Output the difference between the two numbers. The difference can obviously not be a negative number!_

```python
while True:
    try:
        day1weight = int(input("what is your weight on day 1 (kg)"))
        day2weight = int(input("what is your weight on day 2 (kg)"))
        if day1weight > 0 and day2weight > 0:
            difference = day1weight - day2weight
            print(f"the difference in weight is {difference} kg")

            
        else:
            print("please enter a valid weight")
            continue
    except ValueError:
        print("please enter a valid number")
```

_The program you just wrote will fail if floating-point numbers are used. This is actually much more likely. Write a version below that
works for floating-point values._

```python
while True:
    try:
        day1weight = float(input("what is your weight on day 1 (kg)"))
        day2weight = float(input("what is your weight on day 2 (kg)"))
        if day1weight > 0 and day2weight > 0:
            difference = day1weight - day2weight
            print(f"the difference in weight is {difference} kg")

            
        else:
            print("please enter a valid weight")
            continue
    except ValueError:
        print("please enter a valid number")
```

_Suppose the values in your last program were not weights, but the number of sparrows visiting a bird table on two consecutive days. What would you need to change? Write that program below. (You can only have whole numbers of sparrows, obviously.)_

```python
while True:
    try:
        day1sparrows = int(input("how many sparrows visited on day 1: "))
        day2sparrows = int(input("how many sparrows visited on day 2: "))

        if day1sparrows > 0 and day2sparrows > 0:
            total = day1sparrows + day2sparrows
            print(f"total sparrow count is {total} sparrows")
        else:
            print("please enter a valid number of sparrows")
            continue

    except ValueError:
        print("please enter a valid number")
```

_When packing kiwi fruit, the **usual number** to put in a punnet is four. Write a program that prompts for the number of kiwi fruit to be packed, and displays the number of full punnets that will result._

_Remember that the number of fruits in a punnet might change, so be sure to make that as easy as possible to change._

```python
kiwi = int(input("how many kiwis are there: "))
pummets = kiwi / 4
print(f"there are {pummets:.0f} pummets from {kiwi} kiwis")
```

_Peaches, on the other hand, are packed into boxes of 24 fruit. Any remaining after this are packed into smaller punnets of 4. Adapt the program above to handle peaches, and this more complex packing strategy._

```python
peaches = int(input("how many peaches are there: "))
boxes = peaches / 24
remainder_peaches = peaches % 24
small_box = remainder_peaches / 4
print(f"there are {boxes:.0f}  boxes and {small_box:.0f} small boxes from {peaches} peaches")
```
